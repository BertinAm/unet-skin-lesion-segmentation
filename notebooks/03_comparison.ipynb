````python
import sys, os; sys.path.insert(0, os.path.abspath('..'))
import pandas as pd, matplotlib.pyplot as plt
df = pd.read_csv('../outputs/comparison.csv')
df
````

````python
ax = df.set_index('arch')[['dice', 'iou', 'recall']].plot.bar(figsize=(9, 5))
ax.set_ylabel('score'); ax.set_title('Architecture comparison (test set)')
ax.set_ylim(0, 1); plt.xticks(rotation=0); plt.tight_layout(); plt.show()
````
